````
#y = a*x**2 + b*x + c
#a,b,c = cisla, hledam takove x, aby y = 0

#program získá vstupy: a,b,c
#spocitame diskriminant D = b**2 - 4*a*c
#podle diskriminantu rozhodnu o poctu korenu
#spocitam koreny

def ziskej_vstupy():
    a = int(input("Zadej koeficient a: "))
    b = int(input("Zadej koeficient b: "))
    c = int(input("Zadej koeficient c: "))
    return a,b,c

def diskriminant(a,b,c):
    return b**2 - 4*a*c

def zjisti_pocet_korenu(D):
    if D > 0:
        return 2
    elif D == 0:
        return 1
    else:
        return 0

def spocti_koreny(a,b,c,D,nkorenu):
    import math
    if nkorenu == 2:
        x1 = (-b + math.sqrt(D)) / (2*a)
        x2 = (-b - math.sqrt(D)) / (2*a)
        return (x1,x2)
    elif nkorenu == 1:
        return -b/(2*a)
    else:
        return None

def main():
    a,b,c = ziskej_vstupy()
    D = diskriminant(a,b,c)
    nkorenu = zjisti_pocet_korenu(D)
    koreny = spocti_koreny(a,b,c,D,nkorenu)
    print("Kořeny jsou: ", koreny)

if __name__== "__main__":
    main()
````

````

#Funkce která bude generovat webovou stránku s tagy a seznamem věcí
# h1, u1, li, style="padding:xpx; background:barvy;"
#napiste program, ktery vygeneruje html seznam (ul nebo ol) kde *args jsou
#polozky seznamu (li) a **kwargs jsou css atributy 




def seznam_generator(*args, **kwargs):
    kwargs_str = ';'.join('{}:{}'.format(k,v) for k,v in kwargs.items())
    print(f'<ul style="{",".join([kwargs_str])}">')
    for arg in args: 
        print("\t<li>"+arg+("</li>"))
    print("</ul>")
    
def main():
    seznam_generator("houska", "pečen", "kočka", padding= "50px", color= "red")
    
if __name__ == "__main__":
    main()
````
